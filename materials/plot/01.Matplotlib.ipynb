## Работа с объектами в интерактивном режиме

````python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.linspace(0, 10, 100)
y = x**0.5
fig1 = plt.figure()
ax1 = fig1.get_axes()[0]
ax1.plot(x, x**0.3)
````

````python
x = np.linspace(0, 10, 100)
y = x**0.5

fig1 = plt.figure()
plt.plot(x,x**0.5)
ax1 = fig1.get_axes()[0]
ax1.plot(x, x**0.3)
print(fig1 == plt.gcf())
````

````python
fig2 = plt.figure()
ax1 = fig1.get_axes()[0]
ax1.plot(x, x**0.3)
plt.plot(x, x**0.9)
print(fig2 == plt.gcf())
````

````python
fig3 = plt.figure()
plt.plot(x, x**0.9)
ax1 = fig1.get_axes()[0]
ax1.plot(x, x**0.3)
print(fig3 == plt.gcf())
````

Работа с фигурами:

    * `pyplot.figure` --- создает только `Figure` ( или возвращает существующий, если передать `num`)
    * `pyplot.subplots`--- создает `Figure` и создает `Axes` в нем.
    * `pyplot.gcf` --- возвращает текущую `Figure`
    * `pyplot.gca` --- возврващает текущую `Axes`
    * `pyplot.sca` --- устанвливает текущие `Axes` и `Figure`
    * `pyplot.close` --- закрытие `Figure`

## Размещение с помощью `pyplot.subplot`

### Явное размещение

````python
def example_plot(ax, fontsize=12):
    ax.plot([1, 2])
    ax.locator_params(nbins=3)
    ax.set_xlabel('x-label', fontsize=fontsize)
    ax.set_ylabel('y-label', fontsize=fontsize)
    ax.set_title('Title', fontsize=fontsize)


fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)
example_plot(ax1)
example_plot(ax2)
example_plot(ax3)
example_plot(ax4)
````

````python
def example_plot(ax, fontsize=12):
    ax.plot([1, 2])
    ax.locator_params(nbins=3)
    ax.set_xlabel('x-label', fontsize=fontsize)
    ax.set_ylabel('y-label', fontsize=fontsize)
    ax.set_title('Title', fontsize=fontsize)


fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)
example_plot(ax1)
example_plot(ax2)
example_plot(ax3)
example_plot(ax4)
plt.tight_layout()
plt.savefig("filename.png", format="png", trasparent=True, dpi=600)
````

### Сохранение в файл
* Лучше сохранять в векторные форматы SVG и PDF
* Если сохраняете в растровый формат, то увеличивайте [DPI](https://ru.wikipedia.org/wiki/Dots_per_inch)
* Делайте фон прозрачный, так проще потом вставить рисунок

### Настройка стиля

* `pyplot.style.use` --- использование встроенных/сторонних стилей
    * Пример стилевого файла:
```
axes.titlesize : 24
axes.labelsize : 20
lines.linewidth : 3
lines.markersize : 10
xtick.labelsize : 16
ytick.labelsize : 16
```
* `matplotlib.rcParams` --- динамическая настройка
    * Пример:
```
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.linestyle'] = '--'
plt.plot(data)
```
* `matplotlibrc` файл
* Настраиваются цвета, форматы, текст, клавиши GUI и т. д.
